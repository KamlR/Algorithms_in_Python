```python
import time
import random
import matplotlib.pyplot as plt
from makeheap import makeheap
from makeheap_n_log_n import makeheap_n_log_n

sizes = range(1, 101)
times_n = []
times_nlogn = []

for size in sizes:
    t1_sum = 0
    t2_sum = 0
    for _ in range(20):
        arr1 = [random.randint(0, 1000) for _ in range(size)]
        arr2 = arr1.copy()
        start = time.perf_counter()
        makeheap(arr1)
        t1_sum += time.perf_counter() - start

        start = time.perf_counter()
        makeheap_n_log_n(arr2)
        t2_sum += time.perf_counter() - start

    times_n.append(t1_sum / 20)
    times_nlogn.append(t2_sum / 20)

plt.figure(figsize=(8,5))
plt.plot(sizes, times_n, label='makeheap (O(N))', linewidth=2)
plt.plot(sizes, times_nlogn, label='makeheap_n_log_n (O(N log N))', linewidth=2)
plt.xlabel('  N')
plt.ylabel('  (.)')
plt.title('   makeheap  makeheap_n_log_n')
plt.legend()
plt.grid(True)
plt.show()
```
